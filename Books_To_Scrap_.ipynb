### 1. About Project
**Scraping (Collecting) all books information from https://books.toscrape.com/ including book name , price , rate and category of book and saving this collected data in csv file (Books.csv) .** 


### 2. Steps of collecting this data :     
     1.   Collecting Books for any category that've only one web page 
     2.   Collecting Books for any category that've many web page
     3.   Collecting all Categories in the website
     4.   Collecting all books in each categories in the website 
     5.   Saving this data in csv file

### Second Example

```python
response = requests.get('https://books.toscrape.com/catalogue/category/books/womens-fiction_9/index.html')
soup = BeautifulSoup(response.text, 'html.parser')

file_name = soup.find('title').get_text().strip().split(' | ')[0]
file_name += '.csv'

rates = {'One': 1, 'Two': 2, 'Three': 3, 'Four': 4, 'Five': 5}

with open(file_name, mode='w', encoding='utf-8') as csv_file:
    fieldnames = ['book_name', 'book_price', 'book_rate']
    writer = csv.DictWriter(csv_file, fieldnames=fieldnames)

    writer.writeheader()
    for article in soup.find_all('article', attrs={'class': 'product_pod'}):
        book_name = article.find('h3').find('a').get('title')
        book_price = float(article.find('p', attrs={'class': 'price_color'}).get_text()[2:])
        book_rate = rates[article.find('p', attrs={'class': 'star-rating'}).get('class')[-1]]
        writer.writerow({'book_name': book_name, 'book_price': book_price, 'book_rate': book_rate})
```

## **4. Step Four :**

1.   Collecting all books in each categories in the website 

2. Scraping the books (book name , price , rate , category of book) for each category and put them into a CSV file

```python
#Print name of book , rate , category and price of all books in travel category
response = requests.get('https://books.toscrape.com/index.html')
soup = BeautifulSoup(response.text, 'html.parser')
all_categories = soup.find('div', {'class' : 'side_categories'}).find_all('li')[1:]
for category in all_categories :
    cateogry_name = category.get_text().strip()
    cateogry_link = 'https://books.toscrape.com/'+ category.find('a').get('href')
    print(f'Category Name : {cateogry_name}')
    print(cateogry_link)
    response = requests.get(cateogry_link)
    soup = BeautifulSoup(response.text, 'html.parser')
    try :
        page_no = int(soup.find('li', {'class':'current'}).get_text().strip()[-1])
        for i in range(1,page_no+1) :
            link = cateogry_link.replace('index.html','') + 'page-' + str(i) +'.html'
            response = requests.get(link)
            soup = BeautifulSoup(response.text, 'html.parser')
            all_books = soup.find_all('article',  attrs = {'class':'product_pod'})
            print('\n Scraping from : \n\n' + '\t' + link + '\n')
            print('---'*40)
            for book in all_books :
                book_name = book.find('h3').get_text().strip()
                rates = {'One' : 1 , 'Two' : 2 , 'Three' : 3 , 'Four' : 4 , 'Five' : 5}
                book_rating = rates[book.find('p', attrs = {'class':'star-rating'}).get('class')[1]]
                book_price = book.find('div', attrs = {'class':'product_price'}).find('p',{'class' :"price_color"})
                book_price = float(book_price.get_text().split('Â£')[1])
                print(f'Name : {book_name} ||  Rating : {book_rating} || Price :{book_price}')
                print('---'*40)
    except :  
        all_books = soup.find_all('article',  attrs = {'class':'product_pod'})
        print('\n Scraping from : \n\n' + '\t' + cateogry_link + '\n')
        for book in all_books :
                book_name = book.find('h3').get_text().strip()
                rates = {'One' : 1 , 'Two' : 2 , 'Three' : 3 , 'Four' : 4 , 'Five' : 5}
                book_rating = rates[book.find('p', attrs = {'class':'star-rating'}).get('class')[1]]
                book_price = book.find('div', attrs = {'class':'product_price'}).find('p',{'class' :"price_color"})
                book_price = float(book_price.get_text().split('Â£')[1])
                print(f'Name : {book_name} ||  Rating : {book_rating} || Price :{book_price}')
                print('---'*40)  
    print('\n---\n'*10)   
```

```python
#Saving this data that is name of book , rate , category and price of all books in travel category in csv file
with open('Books.csv', mode='w', encoding='utf-8') as csv_file:
    fieldnames = ['Book Name', 'Price', 'Rate','Category']
    writer = csv.DictWriter(csv_file, fieldnames=fieldnames)
    writer.writeheader()    
    response = requests.get('https://books.toscrape.com/index.html')
    soup = BeautifulSoup(response.text, 'html.parser')
    all_categories = soup.find('div', {'class' : 'side_categories'}).find_all('li')[1:]
    for category in all_categories :
        cateogry_name = category.get_text().strip()
        cateogry_link = 'https://books.toscrape.com/'+ category.find('a').get('href')
        print(f'Category Name : {cateogry_name}')
        print(cateogry_link)
        response = requests.get(cateogry_link)
        soup = BeautifulSoup(response.text, 'html.parser')
        try :
            page_no = int(soup.find('li', {'class':'current'}).get_text().strip()[-1])
            for i in range(1,page_no+1) :
                link = cateogry_link.replace('index.html','') + 'page-' + str(i) +'.html'
                response = requests.get(link)
                soup = BeautifulSoup(response.text, 'html.parser')
                all_books = soup.find_all('article',  attrs = {'class':'product_pod'})
                print('\n Scraping from : \n\n' + '\t' + link + '\n')
                print('---'*40)
                for book in all_books :
                    book_name = book.find('h3').get_text().strip()
                    rates = {'One' : 1 , 'Two' : 2 , 'Three' : 3 , 'Four' : 4 , 'Five' : 5}
                    book_rating = rates[book.find('p', attrs = {'class':'star-rating'}).get('class')[1]]
                    book_price = book.find('div', attrs = {'class':'product_price'}).find('p',{'class' :"price_color"})
                    book_price = float(book_price.get_text().split('Â£')[1])
                    writer.writerow({'Book Name': book_name, 'Price': book_price, 
                                     'Rate': book_rating , 'Category' : cateogry_name})
        except :  
            all_books = soup.find_all('article',  attrs = {'class':'product_pod'})
            print('\n Scraping from : \n\n' + '\t' + cateogry_link + '\n')
            print('---'*40)                
            for book in all_books :
                book_name = book.find('h3').get_text().strip()
                rates = {'One' : 1 , 'Two' : 2 , 'Three' : 3 , 'Four' : 4 , 'Five' : 5}
                book_rating = rates[book.find('p', attrs = {'class':'star-rating'}).get('class')[1]]
                book_price = book.find('div', attrs = {'class':'product_price'}).find('p',{'class' :"price_color"})
                book_price = float(book_price.get_text().split('Â£')[1])
                writer.writerow({'Book Name': book_name, 'Price': book_price, 
                                     'Rate': book_rating , 'Category' : cateogry_name})      
        print('\n---\n'*10)                       
```
